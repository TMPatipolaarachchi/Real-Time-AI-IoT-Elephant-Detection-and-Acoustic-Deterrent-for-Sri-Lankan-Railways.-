```python
model = YOLO("/content/drive/MyDrive/VivaDemo/best.pt")
```

```python
results = model.predict(
    source="/content/drive/MyDrive/VivaDemo/images",
    conf=0.55,
    save=True
)
print("Saved to:", results[0].save_dir)
```

```python
names = results[0].names  # class names dict from model

def social_structure(adults, calves):
    total = adults + calves
    if total == 0: return "None"
    if adults == 1 and calves == 0: return "Single"
    if adults >= 2 and calves == 0: return "Herd (Adults only)"
    if adults >= 1 and calves >= 1: return "Herd (Adults + Calf)"
    return "Unknown"

for r in results:
    adults = calves = 0
    for c in r.boxes.cls.tolist():
        label = names[int(c)]
        if "Adult" in label: adults += 1
        elif "Calf" in label: calves += 1

    total = adults + calves
    print({
        "file": r.path.split("/")[-1],
        "elephants_detected": total > 0,
        "adult_count": adults,
        "calf_count": calves,
        "total_elephants": total,
        "social_structure": social_structure(adults, calves)
    })
```

```python
video_in = "/content/drive/MyDrive/VivaDemo/video/test01.mp4"

vid_results = model.predict(
    source=video_in,
    conf=0.35,
    save=True
)
print("Saved video to:", vid_results[0].save_dir)
```

```python
from google.colab import files
import glob

out_dir = str(vid_results[0].save_dir)
out_video = glob.glob(out_dir + "/*.avi")[0]
files.download(out_video)
```

```python
import numpy as np
from collections import Counter

# --- SETTINGS (tune these) ---
BOX_CONF_MIN = 0.25   # ignore weak boxes when counting
FRAME_KEEP_MIN = 1    # ignore frames with 0 elephants
USE_STAT = "median"   # "median" (recommended) or "mode"

names = model.names  # dict like {0:'adult', 1:'calf'}

def social_structure(adults, calves):
    total = adults + calves
    if total == 0: return "None"
    if adults == 1 and calves == 0: return "Single"
    if adults >= 2 and calves == 0: return "Herd (Adults only)"
    if adults >= 1 and calves >= 1: return "Herd (Adults + Calf)"
    return "Unknown"

adult_counts = []
calf_counts = []

for r in vid_results:
    adults = calves = 0

    if r.boxes is not None and len(r.boxes) > 0:
        cls_list = r.boxes.cls.cpu().numpy().astype(int)
        conf_list = r.boxes.conf.cpu().numpy()

        for cls_id, cf in zip(cls_list, conf_list):
            if cf < BOX_CONF_MIN:
                continue  # skip low-confidence boxes

            label = str(names[cls_id]).lower()
            if "adult" in label:
                adults += 1
            elif "calf" in label:
                calves += 1

    # keep only frames with elephants (prevents 0 dominating the statistic)
    if (adults + calves) >= FRAME_KEEP_MIN:
        adult_counts.append(adults)
        calf_counts.append(calves)

# If no frames had detections
if len(adult_counts) == 0:
    final_adults, final_calves = 0, 0
else:
    if USE_STAT == "mode":
        final_adults = Counter(adult_counts).most_common(1)[0][0]
        final_calves = Counter(calf_counts).most_common(1)[0][0]
    else:
        # ✅ Median is robust against occasional false extra detections
        final_adults = int(np.median(adult_counts))
        final_calves = int(np.median(calf_counts))

total = final_adults + final_calves

summary = {
    "elephants_detected": total > 0,
    "adult_count": final_adults,
    "calf_count": final_calves,
    "total_elephants": total,
    "social_structure": social_structure(final_adults, final_calves),
    "debug": {
        "frames_used": len(adult_counts),
        "adult_counts_sample": adult_counts[:10],
        "calf_counts_sample": calf_counts[:10],
        "box_conf_min": BOX_CONF_MIN
    }
}

print(summary)
```
